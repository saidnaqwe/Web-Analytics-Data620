### Why I Chose This Dataset

```python
#package import
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
from operator import itemgetter

import warnings;
warnings.filterwarnings('ignore');
```

```python
#Data Import and statistics
df = pd.read_csv('/content/tute1.csv.csv')
```

```python
# Created a graph from the DataFrame
G = nx.Graph()

# Nodes and edges to the graph
for index, row in df.iterrows():
    G.add_node(row['Quarter'], sales=row['Sales'], ad_budget=row['AdBudget'], gdp=row['GDP'])
    if index > 0:
        G.add_edge(df.iloc[index-1]['Quarter'], row['Quarter'])

# The diameter of the graph
diameter = nx.diameter(G)

# Calculate the average degree of the graph
average_degree = sum(dict(G.degree()).values()) / len(G.nodes())

# Print the results
print(f"Diameter of the graph: {diameter}")
print(f"Average degree of the graph: {average_degree}")

# Draw the graph
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=500, node_color='skyblue', font_size=10, font_weight='bold')

# Created a graph from the DataFrame
G = nx.Graph()

# I Added nodes and edges to the graph
for index, row in df.iterrows():
    G.add_node(row['Quarter'], sales=row['Sales'], ad_budget=row['AdBudget'], gdp=row['GDP'])
    if index > 0:
        G.add_edge(df.iloc[index-1]['Quarter'], row['Quarter'])

# Calculate the diameter of the graph
try:
    diameter = nx.diameter(G)
except nx.NetworkXError:
    diameter = "Undefined (graph is not connected)"

# Calculate the average degree of the graph
average_degree = sum(dict(G.degree()).values()) / len(G.nodes())

# Print the results
print(f"Diameter of the graph: {diameter}")
print(f"Average degree of the graph: {average_degree}")

# Draw the graph
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=500, node_color='skyblue', font_size=10, font_weight='bold')
plt.title('Graph Visualization')
plt.show()
```

### Why I Picked These Graph Layouts

When I'm working with graphs, the layout I choose can really change how I see the data. Heres a quick and easy rundown of some popular layouts and why I think theyre awesome:

1. **Spring Layout**:
   - **Why I Picked It**: I love the spring layout because it makes the graph look balanced and nice. Its like the nodes are magnets pushing away from each other, but the edges are like springs pulling them together.
   - **Details**: This layout spreads out the nodes evenly, so its easier to see the whole structure and how everythings connected. Perfect for when I want a clear and good-looking graph.

2. **Circular Layout**:
   - **Why I Picked It**: The circular layout is great for showing cycles or just making the graph look neat and organized.
   - **Details**: It arranges all the nodes in a circle, which makes it easy to see connections in a tidy way. Great for when I want to highlight the cyclical nature of my data.

3. **Random Layout**:
   - **Why I Picked It**: The random layout isnt super informative, but its handy for testing or debugging.
   - **Details**: It places nodes randomly, which helps me quickly check the basic structure of the graph. Good for initial explorations.

4. **Shell Layout**:
   - **Why I Picked It**: The shell layout is perfect for hierarchical or layered structures.
   - **Details**: Nodes are arranged in concentric circles (shells), which is great for visualizing different levels or groups within the graph.

5. **Spectral Layout**:
   - **Why I Picked It**: The spectral layout is useful for understanding the overall structure of the graph.
   - **Details**: It uses some fancy math to position nodes, helping me see the global structure and identify clusters or communities.

6. **Kamada-Kawai Layout**:
   - **Why I Picked It**: Similar to the spring layout, but it aims to be more precise, resulting in a more uniform distribution of nodes.
   - **Details**: This layout often produces a more evenly spaced graph, which is great for detailed visualizations where uniformity is important.

7. **Planar Layout**:
   - **Why I Picked It**: The planar layout is chosen for graphs that need to avoid edge crossings.
   - **Details**: It positions nodes to avoid edge crossings, making the graph clear and uncluttered. Perfect for planar graphs.

### Conclusion

Each layout has its strengths and I choose them based on what I need for my visualization. The spring layout is great for general-purpose visualizations, the circular layout is perfect for highlighting cycles, and the shell layout is ideal for hierarchical structures. The spectral and Kamada-Kawai layouts give insights into the overall structure, while the planar layout keeps things clear and uncluttered.
